```python
import matplotlib.pyplot as plt

import pandas as pd

wktime1 = pd.read_csv('wktime.T001.his', delimiter = ' ', header= None) #open file and save variables as a dataframe
plt.plot(wktime1[0], wktime1[1], color= 'c', marker= 'o', label= 'wktime1') #label of x and y axis

wktime4 = pd.read_csv('wktime.T004.his', delimiter= ' ', header= None)
plt.plot(wktime4[0], wktime4[1], color='b', marker='*', label='wktime4')
plt.xlabel('W') #x label
plt.ylabel('Probablity p(w)') #y label
plt.title("Probability distribution")#Plot title

plt.xscale("log") #logarithmic scale
plt.yscale("log")
plt.xlim(0.001,100) #range in x
plt.ylim(0.000001,100) #range in y
plt.show()
```
